```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#–ü—Ä–∏–∫–ª–∞–¥ –≤–∏–ºi—Ä—è–Ω–∏—Ö –¥–∞–Ω–∏—Ö i–∑ —Å–µ–Ω—Å–æ—Ä–∞ (ùë•ùë†) —Ç–∞ –µ—Ç–∞–ª–æ–Ω–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω—å (ùë•ùëü)
sensor_readings = np.array([19.8, 24.5, 31.0, 36.2, 41.7]) # Data
  #from sensor
reference_values = np.array([20, 25, 30, 35, 40]) # Reference values
```

```python
#–í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –º–µ—Ç–æ–¥—É –ªi–Ωi–π–Ω–æ—ó —Ä–µ–≥—Ä–µ—Åi—ó –¥–ª—è –∑–Ω–∞—Ö–æ–¥–∂–µ–Ω–Ω—è –∫–æ–µ—Ñi—Üi—î–Ω—Çi–≤ ùëò i b
model = LinearRegression()
model.fit(sensor_readings.reshape(-1, 1), reference_values)
k = model.coef_[0]
b = model.intercept_
print(f"Equation of calibration curve: y = {k:.3f}x + {b:.3f}")
```

```python
# –íi–∑—É–∞–ªi–∑–∞—Üi—è –∫–∞–ªi–±—Ä—É–≤–∞–ª—å–Ω–æ—ó –∫—Ä–∏–≤–æ—ó
plt.figure(figsize=(8, 5))
plt.scatter(sensor_readings, reference_values, label="Calibration data", color="blue")
plt.plot(sensor_readings, model.predict(sensor_readings.reshape(-1,1)), color="red", label="Calibration curve")
plt.xlabel("Sensor values (C)")
plt.ylabel("Reference values (C)")
plt.grid()
plt.legend()
plt.title("Temperature sensor calibration")
plt.show()
```

```python
#–ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –º–æ–¥–µ–ª—ñ
#–í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –∫–∞–ªi–±—Ä—É–≤–∞–ª—å–Ω–æ—ó –º–æ–¥–µ–ªi –¥–ª—è –ø–µ—Ä–µ–≤i—Ä–∫–∏ –Ω–æ–≤–∏—Ö –¥–∞–Ω–∏—Ö
test_readings = np.array([23.0, 38.0, 45.5])
calibrated_values = model.predict(test_readings.reshape(-1, 1))
print("Corrected values:", calibrated_values)
```
