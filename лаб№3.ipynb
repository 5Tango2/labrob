```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs_analog = 5000  # Частота оригинального сигнала (Гц)
t_analog = np.linspace(0, 1, int(fs_analog), endpoint=False)
analog_signal = np.sin(2 * np.pi * 50 * t_analog)  # Частота 50 Гц
```

```python
fs_digital = 500 # Sample rate (Hz)
t_digital = np.linspace(0, 1, int(fs_digital), endpoint=False)
digital_signal = np.sin(2 * np.pi * 50 * t_digital)
```

```python
n_bits = 8 # Amount of bits
levels = 2 ** n_bits
quantized_signal = np.round(digital_signal * (levels // 2)) /(levels // 2)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t_analog, analog_signal, label="Analog signal")
plt.stem(t_digital, digital_signal, linefmt='r-', markerfmt='ro', label="Discrete signal")
plt.step(t_digital, quantized_signal, label="Quantized signal", color="green")
plt.legend()
plt.grid()
plt.show()
```
