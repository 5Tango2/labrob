```python
import numpy as np
import matplotlib.pyplot as plt

# Параметри сигналу
frequency = 4000  # Частота, Гц
duty_cycle = 0.5  # Скважність
amplitude = 7     # Амплітуда
sampling_rate = 100000  # Частота дискретизації, Гц (має бути значно вищою за частоту сигналу)
duration = 0.002   # Тривалість сигналу, с (2 мс)

# Часовий вектор
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Генерація прямокутного сигналу
signal = amplitude * ((t % (1 / frequency)) < (duty_cycle / frequency)).astype(float)

# Візуалізація сигналу
plt.figure(figsize=(10, 4))
plt.plot(t, signal, label="Прямокутний сигнал")
plt.title("Часова форма сигналу")
plt.xlabel("Час, с")
plt.ylabel("Амплітуда")
plt.grid()
plt.legend()
plt.show()

# Дискретне перетворення Фур'є (DFT)
N = len(signal)  # Кількість точок для DFT
dft_result = np.fft.fft(signal)  # Розрахунок DFT
frequencies = np.fft.fftfreq(N, 1 / sampling_rate)  # Частоти

# Візуалізація спектру
plt.figure(figsize=(10, 4))
plt.plot(frequencies[:N // 2], np.abs(dft_result[:N // 2]) / N, label="Амплітуда спектру")
plt.title("Спектр сигналу")
plt.xlabel("Частота, Гц")
plt.ylabel("Амплітуда")
plt.grid()
plt.legend()
plt.show()
```
