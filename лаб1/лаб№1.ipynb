```python
import numpy as np
import matplotlib.pyplot as plt

#  
frequency = 4000  # , 
duty_cycle = 0.5  # 
amplitude = 7     # 
sampling_rate = 100000  #  ,  (      )
duration = 0.002   #  ,  (2 )

#  
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

#   
signal = amplitude * ((t % (1 / frequency)) < (duty_cycle / frequency)).astype(float)

#  
plt.figure(figsize=(10, 4))
plt.plot(t, signal, label=" ")
plt.title("  ")
plt.xlabel(", ")
plt.ylabel("")
plt.grid()
plt.legend()
plt.show()

#   ' (DFT)
N = len(signal)  #    DFT
dft_result = np.fft.fft(signal)  #  DFT
frequencies = np.fft.fftfreq(N, 1 / sampling_rate)  # 

#  
plt.figure(figsize=(10, 4))
plt.plot(frequencies[:N // 2], np.abs(dft_result[:N // 2]) / N, label=" ")
plt.title(" ")
plt.xlabel(", ")
plt.ylabel("")
plt.grid()
plt.legend()
plt.show()
```
